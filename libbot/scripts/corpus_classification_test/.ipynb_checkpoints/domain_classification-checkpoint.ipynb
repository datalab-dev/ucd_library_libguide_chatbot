# Academic Domain Classification of UC Library Research Guides

**LIN 127 Final Project**


I wanted to investigate whether the **linguistic content of chunked library research guide pages** carries enough signal to identify the broad academic domain of the guide it came from, without relying on the guide's title or URL.

The corpus consists of ~7,400 text chunks scraped from UC Davis Library research guides. Each chunk belongs to a specific guide (e.g. *Sociology*, *Agriculture & Related Fields*, *ANS 134: Animal Nutrition Laboratory*), and guides are grouped into four broad academic domains:

| Domain | Examples |
|---|---|
| STEM | Biology, Chemistry, Agriculture, Animal Science, Nutrition |
| Humanities | Literature, History, Art, African American Studies, Film |
| Social Sciences | Sociology, Psychology, Political Science, Economics |
| Health Sciences | Nursing, Medicine, Veterinary, Public Health, Kinesiology |

---

*Do research guides across these domains exhibit measurably different linguistic profiles in how they describe databases, resources, and research tools?*

**Method:** Rule-based domain labeling from guide titles -> NLTK bag-of-words feature extraction -> Naive Bayes classification → Most Informative Features analysis.


## 1. Setup

```python
import pandas as pd
import numpy as np
import re
import random
import warnings
from collections import Counter
warnings.filterwarnings('ignore')

import nltk
from nltk.classify import NaiveBayesClassifier
from nltk.classify.util import accuracy as nltk_accuracy
from nltk.corpus import stopwords

from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

# NLTK data, like in assignment 6
nltk.download('punkt') # word tokenizer model
nltk.download('punkt_tab') # word tokenizer model
nltk.download('averaged_perceptron_tagger') # part-of-speech tagger
nltk.download('averaged_perceptron_tagger_eng') # part-of-speech tagger
nltk.download('stopwords') # stop words list to help filtering out common words


# Reproducibility
RANDOM_SEED = 42
random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)

# ── Configuration ──────────────────────────────────────────────────────────────
DATA_PATH    = '../../../../data/combined_text_full_libguide.csv'   # purposefully vague for path anonimity 
TEST_SPLIT   = 0.20                         # 80/20 train/test split
TOP_WORDS    = 2000                         # vocabulary size for BoW features
```

## 2. Data Loading & Inspection

We load the CSV corpus and take a first look at its shape and field structure.


```python
df = pd.read_csv(DATA_PATH)

print(f"Total chunks: {len(df):,}")
print(f"Columns: {list(df.columns)}")
print(f"Unique guides (parent_id): {df['parent_id'].nunique():,}")
print(f"Unique libguide_titles: {df['libguide_title'].nunique():,}")
print()
df.head(3)
```

### 2a. Constructing `combined_text`

If your CSV already contains a `combined_text` column we use it directly. Otherwise we construct it from the guide title, section title, and chunk body — this gives the classifier access to title-level vocabulary which is highly domain-indicative.


```python
if 'combined_text' not in df.columns:
    df['combined_text'] = (
        'Guide Title: '   + df['libguide_title'].fillna('') + '\n' +
        'Section Title: ' + df['chunk_title'].fillna('')    + '\n' +
                            df['text'].fillna('')
    )
    print("Constructed combined_text from component fields.")
else:
    print("combined_text column already present — using as-is.")

print(f"\nSample combined_text:\n{df['combined_text'].iloc[0][:400]}...")
```

## 3. Domain Labeling

Rather than manually annotating thousands of chunks, we exploit the structured `libguide_title` field as a **free supervision signal**. A keyword-based rule assigns each guide to one of four academic domains.

This is a deliberate methodological choice: the labels come from institutional metadata, not from the text itself, which means the classification task is a genuine test of whether the text *independently* reflects domain membership.

We first strip course-code prefixes (e.g. `ANS 134:`, `NUT 117:`) so that keyword matching operates on the meaningful part of the title.


```python
def clean_title(title: str) -> str:
    """Remove leading course-code prefixes like 'ANS 134: ' or 'NUT 117B: '."""
    return re.sub(r'^[A-Z]{2,4}\s*\d+[A-Z]?\s*:\s*', '', str(title)).strip()


def assign_domain(title: str) -> str:
    """
    Map a libguide_title to one of four broad academic domains using
    keyword matching on the cleaned title.
    Returns 'Unclassified' if no keywords match — these are reviewed separately.
    """
    t = clean_title(title).lower()

    health_kw = [
        'nursing', 'medicine', 'medical', 'pharmacy', 'health', 'clinical',
        'veterinary', 'physician', 'dental', 'epidemiology', 'kinesiology',
        'exercise biology', 'rehabilitation', 'occupational therapy',
        'physical therapy', 'public health', 'physician assistant'
    ]
    stem_kw = [
        'biology', 'biochem', 'chemistry', 'physics', 'mathematics', 'math',
        'statistic', 'engineering', 'computer science', 'agriculture',
        'animal science', 'animal nutrition', 'animal biology',
        'nutrition', 'environmental', 'ecology', 'geology', 'neuroscience',
        'molecular', 'entomology', 'botany', 'plant', 'soil science',
        'food science', 'viticulture', 'enology', 'atmospheric',
        'genomics', 'bioinformatics', 'toxicology', 'microbiology',
        'biological', 'chemical', 'natural resource', 'wildlife',
        'fisheries', 'hydrology', 'climate', 'data science', 'science'
    ]
    humanities_kw = [
        'literature', 'history', 'art history', 'studio art', 'music',
        'philosophy', 'film', 'theater', 'theatre', 'language studies',
        'african american', 'african and african', 'asian american',
        'chicano', 'ethnic studies', 'gender studies', "women's studies",
        'culture', 'classics', 'medieval', 'religious studies', 'religion',
        'archaeology', 'linguistics', 'writing', 'rhetoric', 'poetry',
        'drama', 'creative writing', 'american studies', 'humanities',
        'visual art', 'performance', 'design history'
    ]
    social_kw = [
        'sociology', 'psychology', 'political science', 'economics',
        'anthropology', 'communication', 'education', 'geography',
        'law', 'legal studies', 'criminology', 'social work',
        'public policy', 'international relations', 'human development',
        'cognitive science', 'labor studies', 'urban studies',
        'global studies', 'international'
    ]

    # Check in specificity order (health before STEM to avoid overlap)
    for kw in health_kw:
        if kw in t:
            return 'Health Sciences'
    for kw in stem_kw:
        if kw in t:
            return 'STEM'
    for kw in humanities_kw:
        if kw in t:
            return 'Humanities'
    for kw in social_kw:
        if kw in t:
            return 'Social Sciences'
    return 'Unclassified'


df['domain'] = df['libguide_title'].apply(assign_domain)

print("Domain distribution across chunks:")
print(df['domain'].value_counts().to_string())
```

### 3a. Reviewing Unclassified Guides

Any guide that received the `Unclassified` label needs a manual look. We print the unique titles so we can decide whether to extend the keyword list, manually reclassify, or drop them from the analysis.


```python
unclassified = df[df['domain'] == 'Unclassified']['libguide_title'].unique()
print(f"{len(unclassified)} guide title(s) not matched by keyword rules:\n")
for title in sorted(unclassified):
    print(f"  '{title}'")

print("\nExtend assign_domain() keyword lists above for any titles that should be classified.")
print("Dropping Unclassified chunks from the analysis for now.")

df = df[df['domain'] != 'Unclassified'].copy()
print(f"\nRemaining chunks: {len(df):,}")
```

### 3b. Label Distribution

We visualize both chunk-level and guide-level label distributions. Class imbalance at the chunk level is expected — STEM has many more guides at a research university than, say, Humanities — but we note it for interpretation.


```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Chunk-level
chunk_counts = df['domain'].value_counts()
axes[0].bar(chunk_counts.index, chunk_counts.values, color='steelblue', edgecolor='white')
axes[0].set_title('Chunks per Domain')
axes[0].set_ylabel('Count')
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))

# Guide-level (unique parent_ids per domain)
guide_counts = df.drop_duplicates('parent_id')['domain'].value_counts()
axes[1].bar(guide_counts.index, guide_counts.values, color='coral', edgecolor='white')
axes[1].set_title('Unique Guides per Domain')
axes[1].set_ylabel('Count')

for ax in axes:
    ax.tick_params(axis='x', rotation=15)

plt.suptitle('Domain Label Distributions', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('label_distribution.png', dpi=150)
plt.show()
```

## 4. Feature Extraction

We use a **bag-of-words** representation: for each chunk, we record which words from a shared vocabulary of the top 2,000 most frequent non-stopword tokens are present.

This is a *presence* feature (True/False), not a frequency count — this is standard for NLTK's Naive Bayes classifier and tends to work well for domain classification where a word's mere occurrence is more diagnostic than how often it appears.

**Design decisions:**
- Stopwords (NLTK English list) are removed — function words like *the*, *is*, *of*   are common to all domains and add noise.
- Tokens shorter than 3 characters are dropped — abbreviations and single letters   are usually noise.
- The vocabulary is computed **from the training set only** to avoid data leakage.


```python
STOP_WORDS = set(stopwords.words('english'))

def tokenize(text: str) -> list[str]:
    """Lowercase, tokenize, remove stopwords and short non-alpha tokens."""
    tokens = nltk.word_tokenize(str(text).lower())
    return [t for t in tokens if t.isalpha() and len(t) > 2 and t not in STOP_WORDS]


def build_vocab(texts, top_n: int = TOP_WORDS) -> set:
    """Return the top_n most frequent tokens across a collection of texts."""
    counter = Counter()
    for text in texts:
        counter.update(tokenize(text))
    return {word for word, _ in counter.most_common(top_n)}


def extract_features(text: str, vocab: set) -> dict:
    """Convert a text string into a binary feature dict over vocab."""
    tokens = set(tokenize(text))
    return {word: (word in tokens) for word in vocab}


print(f"Tokenizer ready. Vocabulary will be built from training data (top {TOP_WORDS} words).")
```

## 5. Train / Test Split

We split **by guide** (`parent_id`), not by individual chunk. This is critical: chunks from the same guide share vocabulary, structure, and topic. Splitting by chunk would leak guide-specific language into the test set, artificially inflating accuracy.

By splitting at the guide level, we ensure the classifier is evaluated on its ability to generalize to **entirely unseen guides** — a stricter and more meaningful test.


```python
# Unique parent_ids, shuffled
unique_parents = df['parent_id'].unique().tolist()
random.shuffle(unique_parents)

split_idx   = int(len(unique_parents) * (1 - TEST_SPLIT))
train_pids  = set(unique_parents[:split_idx])
test_pids   = set(unique_parents[split_idx:])

train_df = df[df['parent_id'].isin(train_pids)].copy()
test_df  = df[df['parent_id'].isin(test_pids)].copy()

print(f"Train: {len(train_df):,} chunks from {len(train_pids):,} guides")
print(f"Test:  {len(test_df):,} chunks from {len(test_pids):,} guides")

# Build vocabulary from training data only
vocab = build_vocab(train_df['combined_text'], top_n=TOP_WORDS)
print(f"\nVocabulary size: {len(vocab):,} tokens")
```

```python
print("Extracting features from training set...")
train_features = [
    (extract_features(row['combined_text'], vocab), row['domain'])
    for _, row in train_df.iterrows()
]

print("Extracting features from test set...")
test_features = [
    (extract_features(row['combined_text'], vocab), row['domain'])
    for _, row in test_df.iterrows()
]

print(f"\nTrain feature sets: {len(train_features):,}")
print(f"Test feature sets:  {len(test_features):,}")
```

## 6. Training the Naive Bayes Classifier

We use NLTK's `NaiveBayesClassifier`. Naive Bayes assumes feature independence — i.e. that each word's presence is conditionally independent given the domain label. This is obviously not true in natural language, but in practice the model performs well for text classification tasks, particularly when the goal is *interpretability* rather than maximum accuracy.

Its key advantage here: the learned conditional probabilities translate directly into a ranked list of the most domain-diagnostic words via `show_most_informative_features()`.


```python
print("Training Naive Bayes classifier...")
classifier = NaiveBayesClassifier.train(train_features)
print("Training complete.")
```

## 7. Evaluation

### 7a. Overall Accuracy


```python
acc = nltk_accuracy(classifier, test_features)
print(f"Test accuracy: {acc:.3f}  ({acc*100:.1f}%)")

# Also compute per-class metrics via sklearn
y_true = [label for _, label in test_features]
y_pred = [classifier.classify(feat) for feat, _ in test_features]

print("\nPer-class classification report:")
print(classification_report(y_true, y_pred))
```

### 7b. Confusion Matrix

The confusion matrix reveals *which* domains the classifier conflates — a linguistically meaningful finding in itself. For instance, if STEM and Health Sciences are frequently confused, that suggests their library resource descriptions share substantial vocabulary (clinical databases cited in both, for example).


```python
labels = sorted(df['domain'].unique())
cm = confusion_matrix(y_true, y_pred, labels=labels)

fig, ax = plt.subplots(figsize=(7, 5))
sns.heatmap(
    cm, annot=True, fmt='d', cmap='Blues',
    xticklabels=labels, yticklabels=labels, ax=ax,
    linewidths=0.5, linecolor='lightgray'
)
ax.set_xlabel('Predicted Domain', fontsize=11)
ax.set_ylabel('True Domain', fontsize=11)
ax.set_title('Confusion Matrix — Naive Bayes Domain Classifier', fontsize=12, fontweight='bold')
plt.xticks(rotation=20, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=150)
plt.show()
```

## 8. Most Informative Features — The Core Linguistic Finding

This is the heart of the CL analysis. `show_most_informative_features()` reports the words whose presence (or absence) most strongly shifts the probability toward a particular domain label.

The **likelihood ratio** shown (e.g. `STEM : Health Sciences = 18.4 : 1.0`) means a chunk containing that word is 18.4× more likely to be STEM than Health Sciences. High ratios signal tight, domain-specific vocabulary; low ratios signal shared language.

We print 40 features to get a rich picture of what distinguishes each domain.


```python
print("=" * 60)
print("TOP 40 MOST INFORMATIVE FEATURES")
print("=" * 60)
classifier.show_most_informative_features(40)
```

### 8a. Per-Domain Most Informative Features

To get a cleaner picture of what vocabulary *defines* each domain, we look at the top features that most strongly predict each class specifically, by examining the classifier's probability distributions directly.


```python
from nltk.probability import DictionaryProbDist

print("Top predictive words per domain (based on feature log-probabilities):\n")

domain_top = {}
for domain in sorted(df['domain'].unique()):
    # For each vocab word, score P(word=True | domain) / P(word=True | other)
    scores = {}
    for word in vocab:
        feat = {w: False for w in vocab}
        feat[word] = True
        probs = classifier.prob_classify(feat)
        scores[word] = probs.prob(domain)
    top = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:15]
    domain_top[domain] = top
    print(f"  {domain}:")
    for word, score in top:
        print(f"    {word:<25} P={score:.4f}")
    print()
```

## 9. Error Analysis

Misclassified chunks are linguistically interesting: they represent cases where a guide's language *crosses domain boundaries*. We sample misclassifications and inspect them for patterns — for example, a Sociology chunk about health policy might be predicted as Health Sciences, which is a meaningful and discussable result.


```python
# Attach predictions back to test_df
test_df = test_df.copy()
test_df['predicted_domain'] = [
    classifier.classify(extract_features(row['combined_text'], vocab))
    for _, row in test_df.iterrows()
]
test_df['correct'] = test_df['domain'] == test_df['predicted_domain']

# Overall error rate
error_df = test_df[~test_df['correct']]
print(f"Misclassified chunks: {len(error_df):,} / {len(test_df):,}  ({len(error_df)/len(test_df)*100:.1f}%)")

# Error breakdown: true → predicted
print("\nError breakdown (true → predicted):")
error_breakdown = (
    error_df.groupby(['domain', 'predicted_domain'])
    .size()
    .reset_index(name='count')
    .sort_values('count', ascending=False)
)
print(error_breakdown.to_string(index=False))
```

```python
# Sample 5 misclassified chunks for qualitative inspection
print("Sample misclassified chunks:\n")
sample_errors = error_df.sample(min(5, len(error_df)), random_state=RANDOM_SEED)

for _, row in sample_errors.iterrows():
    print(f"Guide:     {row['libguide_title']}")
    print(f"Section:   {row['chunk_title']}")
    print(f"True:      {row['domain']}  |  Predicted: {row['predicted_domain']}")
    print(f"Text:      {str(row['text'])[:200]}...")
    print("-" * 70)
```

## 10. Saving Results

We append predicted domain labels back to the full corpus and save as CSV. These labels are directly usable as metadata in the original RAG pipeline — for example, to weight retrieval by domain relevance or to filter by subject area.


```python
# Predict on ALL chunks (not just test set) for the output file
print("Generating predictions for full corpus...")
df['predicted_domain'] = [
    classifier.classify(extract_features(row['combined_text'], vocab))
    for _, row in df.iterrows()
]

# Save
output_path = 'corpus_with_domains.csv'
df[['local_id', 'parent_id', 'libguide_title', 'chunk_title',
    'domain', 'predicted_domain', 'text']].to_csv(output_path, index=False)

print(f"Saved to {output_path}")
print(f"\nFinal domain distribution (predicted):")
print(df['predicted_domain'].value_counts().to_string())
```

## 11. Discussion & Conclusion

### What we found

The classifier's accuracy and — more importantly — its most informative features reveal that **academic domain membership is linguistically encoded in the way librarians describe research resources**, even at the level of short text chunks.

Key observations from the most informative features:

- **STEM** guides use precise disciplinary terminology: taxonomic names, compound nouns,   methodology vocabulary (e.g. *reagent*, *genomic*, *protocol*, *specimen*).
- **Humanities** guides favor proper nouns (author names, publication titles),   periodization markers, and interpretive vocabulary (*archive*, *manuscript*, *periodical*).
- **Social Sciences** guides occupy a middle ground, with both quantitative (*survey*,   *dataset*, *longitudinal*) and discursive (*policy*, *discourse*, *qualitative*) terms.
- **Health Sciences** guides are dominated by clinical nomenclature and evidence-based   practice vocabulary (*clinical*, *evidence-based*, *guidelines*, *dosage*).

### Limitations

1. **Label noise:** The keyword-based labeler is imperfect. Some guides — particularly    interdisciplinary ones — may be assigned to the wrong domain.
2. **Class imbalance:** STEM guides likely outnumber Humanities guides significantly    at a research university, which can bias the classifier.
3. **Naive Bayes independence assumption:** Co-occurrence patterns (e.g. *animal* +    *nutrition* together being more diagnostic than either alone) are not captured.
4. **Guide-level vs. chunk-level granularity:** Some chunks (especially short ones    listing a single database) may be too brief to carry reliable domain signal.

### Broader implications for the RAG system

The predicted domain labels produced here are directly usable as metadata in the parent RAG pipeline. Potential applications include:
- **Domain-filtered retrieval:** restrict results to guides in a relevant domain   when the user's query implies a subject area.
- **Reranking:** weight retrieved chunks higher when their predicted domain aligns   with the query's domain.
- **Corpus audit:** identify chunks where the predicted and rule-based labels disagree —   these may be genuinely interdisciplinary resources worth flagging.
